### PostgreSQL (or Postgres) is a popular open-source database system used to store and manage data.

**Environment Variables?**
- Environment variables are key-value pairs used to store configuration settings for applications. Instead of writing these values directly in your code (which can be risky, especially for passwords), they are stored separately to make applications more secure and flexible.


### **Understanding Each Line**

    POSTGRES_HOST="localhost"
- This specifies where the PostgreSQL server is running.
- "localhost" means it is running on the same machine where the application is running. If the database was on another server, you’d use its IP address or domain name instead.
---
    POSTGRES_USER="postgres"
- This defines the username used to connect to the database.
- "postgres" is the default administrative user for PostgreSQL.
---
    POSTGRES_PASSWORD="your_supersafe_password"
- This is the password for the PostgreSQL user.
- It is critical to change this to a secure password before using it in a real system.
---
    POSTGRES_DB="myh_db"
- This sets the name of the database that the application will connect to.
- "myh_db" is just a placeholder name; it can be anything you choose.
---
    POSTGRES_PORT=5432
- This is the port on which PostgreSQL is running.
- The default port for PostgreSQL is 5432, but it can be changed if needed.

# MYH excel data into postgres

```python
from sqlalchemy import create_engine
from dotenv import load_dotenv
import os

load_dotenv()

db_host = os.getenv("POSTGRES_HOST")
db_user = os.getenv("POSTGRES_USER")
db_port = os.getenv("POSTGRES_PORT")
db_password = os.getenv("POSTGRES_PASSWORD")
db_name = os.getenv("POSTGRES_DB")

postgres_connection = create_engine(
    f"postgresql+psycopg2://{db_user}:{db_password}@{db_host}:{db_port}/{db_name}"
)

postgres_connection
```
